```python
import numpy as np
import matplotlib.pyplot as plt
#  y = sin(x)
# 
x = np.array([i*np.pi/180 for i in range(60, 300, 2)])
y = np.sin(x)
# Xy
X = np.matrix(np.column_stack((np.ones(len(x)), x)))
# print(X)
y = np.matrix(y).T
# 
w = (((X.T)*X).I)*(X.T)*y
# 
plt.plot(x, y, 'b', label='data')
plt.plot(x, X*w, 'r', label='fitting curve')
plt.legend()
plt.show()

# 
# thetaalpha
theta = np.matrix(np.zeros((2, 1)))
alpha = 0.001
iteration = 1000
for i in range(iteration):
    gradient = (X.T)*(X*theta - y)
    theta = theta - alpha*gradient
# 
plt.plot(x, y, 'b', label='data')
plt.plot(x, X*theta, 'r', label='fitting curve')
plt.legend()
plt.show()
```
